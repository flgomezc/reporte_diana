# Programa para procesar datos de asistencias en Python

```python
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'

import glob
```

```python
def genera_reporte_en_carpeta(carpeta):
    filenames = glob.glob(carpeta + "/*.xlsx")
    DF_Nino_a_Nino = pd.DataFrame()
    DF_cualitativo = pd.DataFrame()
    
    f = open( carpeta + "/log.log", 'w')
    
    for filename in filenames:
        excelfile = pd.ExcelFile(filename)
        num_sheets = len(excelfile.sheet_names)

        f.write("\nLeyendo archivo: " + filename)
        f.write('\n')
        
        print("\n")
        for j in range(num_sheets):
            f.write("\tLeyendo hoja # " + str(j + 1) + " de " + str(num_sheets)+ " en " + filename)
            f.write('\n')
            df = pd.read_excel(filename, sheet_name=j)

            
            
            entidad = 0
            escenario = 0
            doc_tipoID = 0
            doc_numID = 0
            grupo = 0

            try:     
                df_fechas = df.iloc[9,  [9,10,11,12,13,14,15,16,17]]
                df_fechas = df_fechas.dropna()
                
                df_fechas = pd.to_datetime(df_fechas) #.dt.date             
                fechas = list(df_fechas.apply(lambda x: x.strftime('%-d/%m/%Y')))
                
                print(fechas)
                
                num_fechas = len(fechas)

                entidad = df.iloc[6,2]
                escenario = df.iloc[4,5]
                doc_tipoID = df.iloc[-2,10]
                doc_numID = df.iloc[-2,11]
                if type(doc_numID) == int:
                    pass
                else:
                    doc_numID = int(doc_numID)
                grupo = df.iloc[6,15]

                df_asist = df.iloc[10:-(5+num_fechas), :]
                df_asist = df_asist.set_index('Unnamed: 0')
                df_asist = df_asist.fillna('')
                
                df_activ = df.iloc[-(3+num_fechas):-3, [2,4] ]
                df_activ['fecha'] = fechas
                df_activ['entidad'] = entidad
                df_activ['grupo'] = grupo
                df_activ.columns = ["actividad", "detalle", "fecha", "entidad", "grupo"]
                df_activ['detalle'] = df_activ['detalle'].apply( lambda x: x.replace('\n', '')  )
                
                DF_cualitativo = DF_cualitativo.append(df_activ[["fecha", "actividad", "detalle", 
                                                                 "entidad", "grupo"]], ignore_index=True)


                for i in range(len(fechas)):
                    fecha = fechas[i]   
                    df = df_asist.iloc[:, [0,1,2,3,4,5,6,7, 8+i ]]
                    df.columns=["primer_apellido", "segundo_apellido", "primer_nombre", "segundo_nombre", 
                                        "tipo_id", "id", "ciclo", "genero", 'asistencia']

                    df["entidad"] = entidad
                    df["fecha_impacto"] = fecha    
                    df['tipo_id_formador'] = doc_tipoID
                    df['id_formador'] = doc_numID
                    df['grupo'] = grupo

                    DF_Nino_a_Nino = DF_Nino_a_Nino.append(df, ignore_index=True)
                f.write("    **OK**")
                f.write('\n')

            except Exception as inst:           
                f.write("\t\tEl mediador tiene un error, modific√≥ el formato.")
                f.write('\n')
                       
    f.close()
    
    DF_Nino_a_Nino.to_excel(carpeta + "/reporte_nan.xls")
    DF_cualitativo.to_excel(carpeta + "/reporte_cualitativo.xls")
```

```python
genera_reporte_en_carpeta("data")
```

```python
carpeta = "privados"

genera_reporte_en_carpeta(carpeta)
```
